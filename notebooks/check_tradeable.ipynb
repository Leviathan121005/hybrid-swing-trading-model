```python
symbols = ["BREN", "BBCA", "BYAN", "TPIA", "AMMN", "BBRI", "BMRI", "DSSA", "PANI", "TLKM",
           "ASII", "BBNI", "ICBP", "DNET", "CUAN", "BRIS", "AMRT", "DCII", "UNTR", "SMMA",
           "BRPT", "ISAT", "GOTO", "CPIN", "ADRO", "HMSP", "UNVR", "INDF", "AADI", "KLBF", 
           "GEMS", "MYOR", "BELI", "MASA", "MTEL", "MSIN", "SUPR", "MBMA", "ADMR", "BRMS",
           "MEGA", "NCKL", "TBIG", "MDIY", "PNBN", "BUMI", "BNGA", "CMRY", "SILO", "MDKA"]

eligible_symbols = []

start_date = "2023-01-01"
split_date = "2024-12-31"
end_date = "2025-12-31"

for symbol in symbols:
    symbol = f"{symbol}.JK"
    try:
        stock_prices = get_stock_prices(symbol, start_date, end_date)
        stock_prices_train, stock_prices_test = split_train_test(stock_prices, split_date)

        if pd.to_datetime("2023-01-02") != stock_prices_train["date"].loc[0]:
            continue

        close = stock_prices_train["close"]
        returns = close.pct_change()

        zero_ratio = (returns == 0).mean()
        flat = returns == 0
        streaks = flat.groupby((flat != flat.shift()).cumsum()).sum()
        max_flat_streak = streaks.max()

        is_tradeable = (
            zero_ratio < 0.2 and
            max_flat_streak < 15
        )

        if is_tradeable:
            eligible_symbols.append(symbol)
    except:
        continue

print(eligible_symbols)
```
